## What is Py-ART?


Py-ART is a Python module for plotting, correcting and analyzing weather radar data. The software has been released on GitHub as open source software under a BSD license. 
Runs on Linux, OS X and Windows.

Py-ART can be used for a variety of tasks from basic plotting to more complex processing pipelines. Specific uses for Py-ART include:

- Reading radar data in a variety of file formats.
- Creating plots and visualization of radar data.
- Correcting radar moments while in antenna coordinates. 
- Mapping data from one or multiple radars onto a Cartesian grid.
- Performing retrivals on this mapped data or on the original radial data.
- Writing radial and Cartesian data to NetCDF files.

# What Py-ART is not?

 - Py-ART is NOT a real-time operations radar system. Other open source software (BALTRAD) does a much better job at this.
 
 
 - Py-ART does not have all the latest and greatest correction and retrival algorithms. 
 
 
 - Py-ART is not the only open source radar software out there.

# 1.0 Reading radar files with Py-ART 

We will examine how to read radar files in Py-ART and how Py-ART stores radar volumes. For this we will be examining a file from the W-band Scanning ARM Cloud Radar [WSACR](http://www.arm.gov/instruments/wsacr). 

```python
# turn on inline plotting
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
## We need to import Py-ART
import pyart
```

```python
# let's read the file
radar = pyart.io.read('PyART_Tutorial/data/sgpwsacrcwrhiC1.a1.20120820.204016.nc')
```

We have just created a Radar object which is the class Py-ART uses to store and process radar volume data.

```python
## if we need help we can try:

help(radar)
```

Let's describe the data in the WSARCR file we have just read

```python
radar.info('compact')
```

From the above information we are looking at a Range Height Indicator or RHI scan. This means for each sweep in the volume the azimuth angle is fixed and the radar samples a number of elevation angles. The information on these elevation angles is stored in the elevation attribute as a Python dictionary.

```python
radar.elevation.keys()
```

```python
print radar.elevation['long_name']
print radar.elevation['standard_name']
print radar.elevation['units']
```

```python
# data
radar.elevation['data']
```

We can plot the elevation angles for all the sweeps in the volume


```python
plt.plot(radar.elevation['data'])
```

# Plotting radar data with Py-ART

```python
#Fields attribute --> a dictionary of dictionaries containing the radar fields or moments
radar.fields.keys()
```

```python
#Lets use RadarDisplay class for plottinig radar data taking into
#account the pointing data.
display = pyart.graph.RadarDisplay(radar)


# We can find a lot of convenience functions which calls the appropriate 
# plotting function depending on the scan type of the radar volume

#here we will use plot_rhi in order to visualize the reflectivity data from the second
#sweep in the WSACR radar volume

display.plot_rhi('reflectivity', sweep=1, vmin=-32, vmax=20)
#vmin and vmax means luminance minimum and maximum values,

```

```python
# We can also perform zoom in
# zooming in on an interesting region oof this plot can be
# accomplished using the SET_LIMITS method

display.plot('reflectivity', sweep=1, vmin=-32, vmax=20)
display.set_limits(ylim=(7,12), xlim=(-18, -10))
```
